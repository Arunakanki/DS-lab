```python
import matplotlib.pyplot as plt
```

```python
hours_studies=[10,9,2,15,10,16,11,16]
exam_scores=[95,80,10,50,45,98,38,93]
```

```python
plt.plot(hours_studies,exam_scores,marker='*',color='red',linestyle='-')
plt.xlabel('Hours studies')
plt.ylabel('Score in finals exam')
plt.title('effect of hours studied on exam scorse')
```

```python
plt.show()
```
